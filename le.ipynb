```python
import pandas as pd
from sqlalchemy import create_engine
```

## Bring in csv files and load into dataframes

```python
#Bring in life expectancy csv
le_file = "Resources/life_expectancy.csv"
life_2010 = pd.read_csv(le_file)
life_2010.head()
```

```python
#Bring in life expectancy csv
le_file = "Resources/life_expectancy.csv"
life_2000 = pd.read_csv(le_file)
life_2000.head()
```

## Cleaning of Data

#### Life Expectancy in 2010

```python
life_2010.head()
```

```python
#just show years of 2000 to 2010
life_2010 = life_2010.loc[life_2010['Year']==2010]
life_2010.head()
```

```python
# Rename life expectancy column
life_2010 = life_2010.rename(columns={"Life expectancy ":"life_expectancy", "Country":"country","Status":"status"})
life_2010.head()
```

```python
#Get only the columns we want
life_2010 = life_2010.drop(life_2010.iloc[:, 4:22], axis = 1) 
life_2010.head()
```

```python
life_2010 = life_2010.drop(columns="Year")
life_2010.head()
```

```python
#confirmation
life_2010.columns
```

### Same process but for 2000

```python
life_2000.head()
```

```python
#Show years only in 2000
life_2000 = life_2000.loc[life_2000['Year']==2000]
life_2000.head()
```

```python
# Rename life expectancy column
life_2000 = life_2000.rename(columns={"Life expectancy ":"life_expectancy", "Country":"country","Status":"status"})
life_2000.head()
```

```python
#Get only the columns we want
life_2000 = life_2000.drop(life_2000.iloc[:, 4:22], axis = 1) 
life_2000.head()
```

```python
life_2000 = life_2000.drop(columns="Year")
life_2000.head()
```

## Create database connection


```python
import getpass
password = getpass.getpass()
```

```python
#Create connection
engine = create_engine(f"postgresql://postgres:{password}@localhost:5432/ETL_db")
```

```python
# See tables
engine.table_names()
```

## Load dataframe into table in database

```python
# Putting data from life_exp dataframe into table in SQL
life_2010.to_sql(name='life_expectancy_2010', con=engine, if_exists='append', index=False)
```

```python
#confirm sql query from table
pd.read_sql_query("select * from life_expectancy_2010", con=engine).head()
```

```python
# Putting data from life_exp dataframe into table in SQL
life_2000.to_sql(name='life_expectancy_2000', con=engine, if_exists='append', index=False)
```

```python
#confirm sql query from table
pd.read_sql_query("select * from life_expectancy_2000", con=engine).head()
```
